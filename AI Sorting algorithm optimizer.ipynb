```python
import zipfile
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Specify the path to your zip file
zip_file_path = "C:/Users/Unnati Yadav/Downloads/archive.zip"

# Define the name of the .txt file to be extracted and cleaned
txt_file_name = None

# Step 1: List all .txt files in the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zipf:
    print("List of .txt files in the archive:")
    for file_info in zipf.infolist():
        if file_info.filename.endswith('.txt'):
            print(file_info.filename)
            if txt_file_name is None:
                txt_file_name = file_info.filename  # Use the first .txt file found

# Verify that a .txt file was selected
if txt_file_name is None:
    raise FileNotFoundError("No .txt file found in the ZIP archive.")

# Define the path where the extracted .txt file will be temporarily stored
extracted_txt_path = "extracted_data.txt"
cleaned_csv_path = "preprocessed_data.csv"

# Step 2: Extract the selected .txt file
with zipfile.ZipFile(zip_file_path, 'r') as zipf:
    zipf.extract(txt_file_name, path='.')
    os.rename(txt_file_name, extracted_txt_path)  # Rename for easier access

try:
    # Step 3: Load the .txt file into a DataFrame (modify the delimiter if necessary)
    data = pd.read_csv(extracted_txt_path, delimiter=',', encoding='utf-8')

    # Step 4: Data Cleaning
    # Remove any duplicate rows
    data = data.drop_duplicates()
    # Drop rows with any missing values
    data = data.dropna()

    # Step 5: Data Transformation
    # Example: Convert categorical columns to numerical using Label Encoding
    label_encoder = LabelEncoder()
    for column in data.select_dtypes(include=['object']).columns:
        data[column] = label_encoder.fit_transform(data[column])

    # Example: Scale numerical columns using Standard Scaling
    scaler = StandardScaler()
    numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
    data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

    # Step 6: Save the preprocessed data to a CSV file
    data.to_csv(cleaned_csv_path, index=False, lineterminator='\n', encoding='utf-8')
    print(f"Data preprocessed and saved to {cleaned_csv_path}")

finally:
    # Clean up by removing the extracted .txt file
    if os.path.exists(extracted_txt_path):
        os.remove(extracted_txt_path)
```

```python
# Step 4: Data Cleaning and Transformation
# Drop duplicates
data = data.drop_duplicates()

# Fill missing numerical values with median
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].median())

# Fill missing categorical values with mode
categorical_columns = data.select_dtypes(include=['object']).columns
data[categorical_columns] = data[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Apply one-hot encoding to categorical columns
data = pd.get_dummies(data, columns=categorical_columns)

# Outlier removal using IQR for numerical columns
for column in numerical_columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    data = data[(data[column] >= (Q1 - 1.5 * IQR)) & (data[column] <= (Q3 + 1.5 * IQR))]

# Scaling
scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

# Save the cleaned and transformed data
data.to_csv(cleaned_csv_path, index=False)
```

```python
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
duplicate_rows = data[data.duplicated()]
print("Duplicate rows remaining:", len(duplicate_rows))

```
